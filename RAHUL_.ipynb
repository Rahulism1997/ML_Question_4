```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler


df = pd.read_csv('/content/WineQT.csv')


print("Initial Data:")
print(df.head())


print("\nRows where quality <= 6:")
print(df[df['quality'] <= 6].head())

print("\nRows where quality > 6:")
print(df[df['quality'] > 6].head())

df.loc[df['quality'] <= 6, 'quality'] = 0
df.loc[df['quality'] > 6, 'quality'] = 1

print("\nData after converting quality to binary:")
print(df.head())

norm = df.columns[df.columns != 'quality']
scaler = MinMaxScaler()
df[norm] = scaler.fit_transform(df[norm])

print("\nNormalized Data:")
print(df.head())
```
